```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/devzohaib/Simple-Linear-Regression/master/tvmarketing.csv"
ds = pd.read_csv(url)
x = ds.TV
y = ds.Sales
plt.scatter(x, y)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("DataSet")
plt.show()
```

### Splitting the data into train and test data 

the convention for converting the ds into train test are first train datas and then train datas

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(x,y,train_size=0.7,random_state=0000)
# print(X_train,Y_train)
print(X_test.shape)
```

# now fixing the array dimension as required for sklearn convention

```python
X_train = X_train[:, np.newaxis]
X_test = X_test[:, np.newaxis]
print(X_test.shape,X_train.shape)
```

# now using linear regression from sklearn

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
print(X_test.shape,X_train.shape)
lr.fit(X_train,Y_train)
```

finding out the coefficients 

```python
print(lr.coef_)
print(lr.intercept_)
```

now finding the predicated value 

```python
Y_pred = lr.predict(X_test)
# print(Y_pred,Y_test)
plt.scatter(X_test,Y_test)
plt.xlabel("Predicated")
plt.ylabel("Actual")
plt.plot(X_test,Y_pred,linestyle="dotted",color="red")
plt.show()


plt.show()
```

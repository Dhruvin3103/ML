```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head(4)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["species"].nunique()
```

```python
df['sex'].nunique()
```

```python
island = pd.get_dummies(data=df["island"],drop_first=True)
df = pd.concat([df,island],axis =1)
df.head(2)
```

```python
sex = pd.get_dummies(data=df["sex"],drop_first=True)
df = pd.concat([df,sex],axis =1)
df.head(2)
```

```python
df.drop(['sex','island'],axis=1,inplace=True)
df.head(2)
```

```python
Y = df["species"]
Y.unique()
Y = Y.map({'Adelie':0,'Gentoo':1,'Chinstrap':2})
Y.head(2)
```

```python
df.drop("species",axis=1,inplace=True)
df.head(2)
```

```python
X = df
X.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=42)
X_train.head(2)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(X_train,Y_train)
```

```python
RF.score(X_test,Y_test)
```

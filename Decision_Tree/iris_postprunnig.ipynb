```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
ir = load_iris()
dir(ir)
```

```python
df = sns.load_dataset("iris")
df.head(5)
```

```python
X = df.iloc[:,[0,1,2,3]]
X.head(5)
```

```python
Y = ir.target
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33,random_state=None)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

### Using withour postprunnig

```python
tree = DecisionTreeClassifier()
tree.fit(X_train,Y_train)
```

```python
tree.score(X_test,Y_test)
```

```python
plt.figure(figsize=(15,10))
plot_tree(tree,filled=True)
```

### Using postprunnig 

```python
treep = DecisionTreeClassifier(max_depth=3)
treep.fit(X_train,Y_train)
```

```python
treep.score(X_test,Y_test)
```

```python
plt.figure(figsize=(15,10))
plot_tree(treep,filled=True)
```

```python
treep.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy : ",accuracy_score(Y_test,treep.predict(X_test)))
print("classifiction : ",classification_report(Y_test,treep.predict(X_test)))
```
